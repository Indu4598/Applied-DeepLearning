## Question 1

```python
def KFoldCV(nums,k):
    random.shuffle(nums)
    num_ele = len(nums)//k
    for i in range(k):
        lb = i*num_ele
        ub = (i+1)*num_ele
        test = nums[lb:ub]
        train = [i for i in nums if i not in test]
        print("Training: ",train,"Testing: ",test)
```

```python
def ConfMatrix(M):
    tp = M[0,0]
    fp = M[0,1]
    fn = M[1,0]
    tn = M[1,1]
    recall = tp/(tp+fn)
    precision = tp/(tp+fp)
    f = 2*recall*precision/(recall + precision)
    ppv = tp/(tp+fp)
    npv = tn/(tn+fn)
    sensitivity = tp/(tp+fn)
    specificity = tn/(tn+fp)
    accuracy = (tp+tn)/(tp+tn+fp+fn)
    print("f1-score: ", f)
    print("positive predicted value: ", ppv)
    print("negative predicted value: ", npv)
    print("sensitivity: ", sensitivity)
    print("specificity: ", specificity)
    print("accuracy: ",accuracy)
```

```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.model_selection import train_test_split,StratifiedKFold,KFold
from sklearn.metrics import f1_score,log_loss,precision_recall_curve, average_precision_score,roc_curve, auc, roc_auc_score,confusion_matrix, classification_report
from sklearn.utils import resample
from sklearn.linear_model import LogisticRegression
from sklearn import svm

import pandas as pd
import xgboost as xgb
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Loading the data
data = sklearn.datasets.load_digits()
x, y = data.data, data.target
```

```python
#Checking the data shape
print (x.shape, y.shape)
```

```python
#Splitting the data into train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30,shuffle=True)
```

```python
# Composition of train and test
print (x_train.shape, y_train.shape)
print (x_test.shape, y_test.shape)
```

```python
#Conerting train and test to DataFrame and Series respectively
x_train = pd.DataFrame(x_train)
y_train = pd.Series(y_train)
x_test = pd.DataFrame(x_test)
y_test = pd.Series(y_test)
```

```python
y_train_5 = y_train==5
y_test_5 = y_test==5
```

```python
y_train_5.replace(True,1,inplace=True)
y_test_5.replace(True,1,inplace=True)
```

```python
# Bootstrap distinct Samples S1,S2,and S3.
S1 = resample(x_test,y_test_5,replace=True,n_samples=len(x_test),random_state=30)
S2 = resample(x_test,y_test_5,replace=True,n_samples=len(x_test),random_state=40)
S3 = resample(x_test,y_test_5,replace=True,n_samples=len(x_test),random_state=50)
```

```python
# x and y of S1,S2,and S3
s1_x,s1_y = S1[0],S1[1]
s2_x,s2_y = S2[0],S2[1]
s3_x,s3_y = S3[0],S3[1]
```

```python
# Splitting the data into 10-folds
k_fold = KFold(n_splits=10,shuffle=True,random_state=209)
```

```python
# Setting up parameters before performing cross validation
trainingScores = []
cvScores = []
model = []
predictionsBasedOnKFolds = pd.DataFrame(data=[],index=y_train.index,columns=[0,1])
model = xgb.XGBClassifier(max_depth=3, n_estimators=300, learning_rate=0.05,random_state=42)
logreg = LogisticRegression(C=1e3,solver='liblinear',random_state=42)
clf = svm.SVC(kernel='linear',C=1.0)
```

```python
def vote(xgb,lr,svm):
    predictions = np.zeros(xgb.shape[0])
    for i in range(xgb.shape[0]):
        vote_list = [xgb[i],lr[i],svm[i]]
        final_vote = max(set(vote_list), key = vote_list.count)
        predictions[i]=final_vote
    return predictions
```

```python
#Performing 10-fold cross validation to check for overfitting and underfitting.
avg_f1_score_xg=0
avg_f1_score_lr=0
avg_f1_score_svm=0
avg_f1_score_vote=0
i=1
print("Fold","","XGB","   ","  lr","   ","  SVM","  vote")
for train_index, cv_index in k_fold.split(np.zeros(len(x_train)) ,y_train_5.ravel()):
    #getting the folds 
    x_train_fold, x_cv_fold = x_train.iloc[train_index,:], x_train.iloc[cv_index,:]
    y_train_fold, y_cv_fold = y_train_5.iloc[train_index], y_train_5.iloc[cv_index]
    
    #fitting the models
    gbm = model.fit(x_train_fold, y_train_fold)
    logreg.fit(x_train_fold, y_train_fold)
    clf.fit(x_train_fold,y_train_fold)
    
    #Predictions 
    predictions_xg = gbm.predict(x_cv_fold)
    predictions_lr = logreg.predict(x_cv_fold)
    predictions_svm = clf.predict(x_cv_fold)
    predictions_vote = vote(predictions_xg,predictions_lr,predictions_svm)

    
    #F1-Score on different models
    f1_score_xg = f1_score(y_cv_fold,predictions_xg)*100
    f1_score_lr = f1_score(y_cv_fold,predictions_lr)*100
    f1_score_svm = f1_score(y_cv_fold,predictions_svm)*100
    f1_score_vote = f1_score(y_cv_fold,predictions_vote)*100
    
    
    print(i,"   "+"{:.3f}".format(f1_score_xg),"  "+"{:.3f}".format(f1_score_lr),"  "+"{:.3f}".format(f1_score_svm),"  "+"{:.3f}".format(f1_score_vote))
    avg_f1_score_xg+= f1_score_xg
    avg_f1_score_lr+= f1_score_lr
    avg_f1_score_svm+=f1_score_svm
    avg_f1_score_vote+=f1_score_vote
    i+=1

print("Average F1-Score XGB:"+" {:.2f}".format(avg_f1_score_xg/10))
print("Average F1-Score lr:"+"{:.2f}".format(avg_f1_score_lr/10))
print("Average F1-Score SVM:"+"{:.2f}".format(avg_f1_score_svm/10))
print("Average F1-Score Voting:"+"{:.2f}".format(avg_f1_score_vote/10))
```

```python
#Fitting the model 
gbm_train = xgb.XGBClassifier(max_depth=3, n_estimators=700, learning_rate=0.05,random_state=42).fit(x_train, y_train_5)
logreg_train = LogisticRegression(C=1e3,solver='liblinear').fit(x_train,y_train_5)
clf_train = svm.SVC(kernel='linear',C=1.0).fit(x_train,y_train_5)
```

```python
# XGB on three bootstraps
predictions_gbm_s1 = gbm_train.predict(s1_x)
f1_gbm_s1 = f1_score(s1_y,predictions_gbm_s1)*100
print("S1 boot strap:",f1_gbm_s1)

predictions_gbm_s2 = gbm_train.predict(s2_x)
f1_gbm_s2 = f1_score(s2_y,predictions_gbm_s2)*100
print("S2 boot strap:",f1_gbm_s2)

predictions_gbm_s3 = gbm_train.predict(s3_x)
f1_gbm_s3 = f1_score(s3_y,predictions_gbm_s3)*100
print("S3 boot strap:",f1_gbm_s3)

Average_f1_gbm = (f1_gbm_s1 + f1_gbm_s2 + f1_gbm_s3 )/3
print("average f1 for gbm:",Average_f1_gbm)
```

```python
# Logistic Regression on three bootstraps
predictions_lr_s1 = logreg_train.predict(s1_x)
f1_lr_s1 = f1_score(s1_y,predictions_lr_s1)*100
print("S1 boot strap:",f1_lr_s1)

predictions_lr_s2 = logreg_train.predict(s2_x)
f1_lr_s2 = f1_score(s2_y,predictions_lr_s2)*100
print("S2 boot strap:",f1_lr_s2)

predictions_lr_s3 = logreg_train.predict(s3_x)
f1_lr_s3 = f1_score(s3_y,predictions_lr_s3)*100
print("S3 boot strap:",f1_lr_s3)

Average_f1_lr = (f1_lr_s1 + f1_lr_s2 + f1_lr_s3 )/3
print("average f1 for lr:",Average_f1_lr)
```

```python
# SVM on three bootstraps
predictions_svm_s1 = clf_train.predict(s1_x)
f1_svm_s1 = f1_score(s1_y,predictions_svm_s1)*100
print("S1 boot strap:",f1_svm_s1)

predictions_svm_s2 = clf_train.predict(s2_x)
f1_svm_s2 = f1_score(s2_y,predictions_svm_s2)*100
print("S1 boot strap:",f1_svm_s2)

predictions_svm_s3 = clf_train.predict(s3_x)
f1_svm_s3 = f1_score(s3_y,predictions_svm_s3)*100
print("S1 boot strap:",f1_svm_s3)

Average_f1_svm = (f1_svm_s1 + f1_svm_s2 + f1_svm_s3 )/3
print("average f1 for svm:",Average_f1_svm)
```

```python
predictions_vote_s1 = vote(predictions_gbm_s1,predictions_lr_s1,predictions_svm_s1)
f1_vote_s1 = f1_score(s1_y,predictions_vote_s1)*100
print("S1 boot strap:",f1_svm_s1)

predictions_vote_s2 = vote(predictions_gbm_s2,predictions_lr_s2,predictions_svm_s2)
f1_vote_s2 = f1_score(s2_y,predictions_vote_s2)*100
print("S1 boot strap:",f1_svm_s2)

predictions_vote_s3 = vote(predictions_gbm_s3,predictions_lr_s3,predictions_svm_s3)
f1_vote_s3 = f1_score(s3_y,predictions_vote_s3)*100
print("S1 boot strap:",f1_svm_s3)

Average_f1_vote = (f1_vote_s1 + f1_vote_s2 + f1_vote_s3 )/3
print("average f1 for svm:",Average_f1_vote)
```

```python
predictions_lr_test = logreg_train.predict(x_test)
precision, recall, thresholds = precision_recall_curve(y_test_5,predictions_lr_test)
```
